### Why External Long-Term Financing?

*   **Internal vs. External Funds:**
    *   Firms first try to fund growth through retained earnings (internal equity).
    
    *   However, ambitious growth plans, large investments (R&D, expansion), or M&A often require more capital than internal funds can provide.

*   **The Need for External Capital:**
    *   Scale Operations: Increase production, enter new markets.
    
    *   Fund R&D: Develop new products/technologies (like MM's next-gen sensor).
    
    *   Strategic Acquisitions: Acquire competitors or complementary businesses.
    
    *   Bridge to Profitability: For early-stage companies yet to achieve positive cash flow.
    
    *   Restructure Balance Sheet: Manage existing debt, optimize capital structure.

*   **Why might a very profitable, mature company still choose to raise external capital instead of solely relying on its retained earnings?**

### Early-Stage Equity - Angel Investors & Venture Capital (VC)

*   **Angel Investors:**
    *   Wealthy individuals investing their own capital in startups.
    
    *   Often former entrepreneurs with industry expertise.
    
    *   Typically invest in very early stages (seed funding).
    
    *   Provide smaller amounts than VCs (e.g., MM's initial angel round).
    
    *   May invest individually or in "angel groups."

    *   What are main the motivations of the "Angel Investors"?

*   **Venture Capital (VC) Firms:**
    *   Organized as limited partnerships; invest institutional money (pension funds, endowments) in high-growth potential private companies.
    
    *   Invest in stages (Series A, B, C...) as company meets milestones.
    
    *   Provide larger capital amounts than angels.
    
    *   Take active role: board seats, strategic guidance, network access.
    
    *   Aim for high returns (e.g., 10x) over 5-7 years, primarily through IPO or acquisition.

### VC Funding Rounds & Valuation

*   **Staged Financing:** VCs invest in "rounds" (Seed, Series A, B, etc.) tied to milestones, reducing their risk.

*   **Valuation is Key:**
    *   Pre-Money Valuation: Value of the company before a new investment.
        *   Pre-Money Value = Post-Money Value - Investment Amount
    
    *   Post-Money Valuation: Value of the company *after* a new investment.
        *   Post-Money Value = Offer Price per Share * Total Shares Outstanding (post-investment)
        
*   **Example (MM's Series A):**
    *   Innovate Ventures invested $10M for 25% equity.
    
    *   Post-Money Valuation = $10M / 0.25 = $40M.
    
    *   Pre-Money Valuation = $40M - $10M = $30M.

*   Valuation methods: Comparables, DCF (challenging for early stage), market conditions.

### VC Terms - Convertible Preferred Stock & Liquidation Preferences

*   **VCs rarely buy common stock directly.** They typically receive "Convertible Preferred Stock".
    *   **Convertible Preferred Stock:**
    
        *   Has features of both debt (fixed liquidation preference) and equity (conversion option).
    
        *   Conversion Option: Holder can convert preferred shares into a predetermined number of common shares, usually when advantageous (e.g., at IPO or high-value acquisition).

*   **Liquidation Preference:**
    *   In case of sale, merger, or liquidation, preferred stockholders get their money back *before* common stockholders.
    
    *   Typically "1x" (get initial investment back). Can be higher (2x, 3x).
    
    *   Non-Participating: Holder chooses *either* liquidation preference OR conversion to common.
    
        *   MM Series A Example (Innovate Ventures): 1x non-participating. If MM sold for $100M, Innovate chooses $10M OR 25% of $100M = $25M. They'd choose $25M.
    
    *   Participating Preferred: Holder gets liquidation preference back *AND* shares in remaining proceeds as if converted (pro-rata with common).
    
        *   Case Note Example: 1x participating on $10M for 25% in a $100M sale: $10M + 25% of ($100M - $10M) = $10M + $22.5M = $32.5M. Highly dilutive to common.

### Strategic Corporate Investors vs. Financial VCs

*   **Financial VCs (like Innovate Ventures):**
    *   Primary Goal: Maximize financial return (IRR, Multiple on Invested Capital) for their Limited Partners (LPs).
    *   Exit Horizon: Typically 3-7 years.
    *   Involvement: Board seat, financial oversight, strategic guidance towards exit.
    *   Terms: Focused on downside protection (liquidation preference) and upside (conversion, anti-dilution).
*   **Strategic Corporate Investors (like the conglomerate considering MM):**
    *   Goals: Can be financial, but often include strategic objectives:
        *   Access to new technology/innovation.
        *   Entry into new markets.
        *   Potential future acquisition target.
        *   Blocking competitors.
    *   Exit Horizon: Can be longer or less defined; may aim for integration/acquisition.
    *   Involvement: Board seat, but also potential operational collaboration, distribution channels, or technology integration.
    *   Terms: May include financial terms similar to VCs, but also "Strategic Terms" like exclusivity, rights of first refusal on technology, favored customer pricing (as in MM case). These have economic value/cost.

### Growth Stage Debt - Venture Debt

*   **Traditional Bank Loans:** Often unsuitable for early/growth-stage tech firms with:
    *   Negative cash flows.
    
    *   Few tangible assets for collateral.
    
    *   High uncertainty.
*   **Venture Debt:**
    *   Specialized debt for venture-backed companies, often provided alongside or between equity rounds.
    
    *   Providers: Specialized banks or non-bank lenders.
    
    *   Purpose: Extend cash runway, fund specific projects, bridge to next equity round, reduce dilution from equity.
    
    *   Key Features:
    
        *   Less dilutive than equity of the same amount *initially*.
    
        *   Interest rates typically higher than bank loans (e.g., MM case: 8-12%).
    
        *   "Equity Kicker" (Warrants): Lenders often receive warrants (options to buy equity at a set price) to compensate for higher risk and allow participation in upside. This adds to the true cost of debt.
    
        *   Covenants: Can be restrictive (e.g., minimum cash balance, performance metrics).
    
        *   Secured by assets, often intellectual property or senior to other claims.
*   Why not Public Bonds for MM?
    *   MM is private, 5 years old, no public credit rating.
    
    *   Public bond market (even high-yield) requires larger scale, longer operating history, public financials. Would demand very high rates & collateral if even accessible.

### Initial Public Offering (IPO) - Process & Purpose

*   **IPO:** First sale of stock by a private company to the public. Shares become traded on a public exchange (e.g., NYSE, Nasdaq).

*   **Primary Purposes:**
    *   Raise significant capital for growth, acquisitions, debt repayment.
    
    *   Provide liquidity for existing shareholders (founders, employees, VCs) to sell their shares.
    
    *   Enhance public profile, credibility, brand awareness.
    
    *   Create publicly traded stock for use in employee compensation or acquisitions.
*   **Key Players & Process Overview:**
    1.  Company & Underwriter(s): Company selects an investment bank (lead underwriter) to manage the IPO. Underwriters form a syndicate to share risk and distribute shares.
    
    2.  Due Diligence & Registration: SEC filings (e.g., S-1 registration statement, prospectus detailing company info, financials, risks).
    
    3.  Valuation & "Roadshow": Underwriters and management market the IPO to institutional investors to gauge interest and refine offer price range.
    
    4.  Pricing & Allocation: Final offer price set, shares allocated to investors.
    
    5.  Trading Begins: Shares start trading on the exchange.
*   **Primary vs. Secondary Offering:**
    *   Primary: New shares issued by company, proceeds go to company.
    
    *   Secondary: Existing shareholders sell their shares, proceeds go to sellers. (IPOs often include both).

### IPO - Key Issues (Underpricing)

*   **Underpricing ("IPO Pop"):**
    *   The offer price set by underwriters is often lower than the price at which the stock first trades in the secondary market.
    
    *   Average first-day returns for IPOs historically have been positive (e.g., MM case note: 15-20% for tech IPOs, but variable).
*   **Why Underprice?**
    *   Winner's Curse: To induce uninformed investors to participate, knowing that if an IPO is "hot" (good), they'll be rationed, and if it's "cold" (bad), they'll get all they asked for. Underpricing compensates for this adverse selection.
    
    *   Information Asymmetry: Underwriters have better information than the average investor. Underpricing reduces their risk of being stuck with unsold shares.
    
    *   Analyst Coverage & Marketing: A "successful" pop generates positive buzz and encourages analyst coverage.
    
    *   Leaving Money on the Table: The company raises less capital than it could have if priced higher. Pre-IPO shareholders experience more dilution for the same capital raised (or less capital for same dilution).

### Impact on Founders/Existing Shareholders - Dilution & Control

*   **All external financing methods have implications for existing shareholders:**

*   **Equity Financing (VC, Strategic, IPO):**
    *  Dilution: Issuing new shares reduces the percentage ownership of existing shareholders.
        
        *   New % Ownership = Old Shares / (Old Shares + New Shares Issued)
    
    *   Control: New equity investors (especially VCs and strategic partners) often gain:
        *   Board seats.
    
        *   Veto rights over major decisions (liquidation preferences also influence control over exit).
    
        *   Information rights.
    
        *   IPO leads to dispersed ownership, scrutiny from public investors and analysts, regulatory compliance burden.

*   Debt Financing (Venture Debt):
    *   Dilution: Less direct upfront dilution than equity. Warrants cause some future potential dilution if exercised.
    
    *   Control: Lenders generally don't take board seats or voting rights. However, debt covenants can restrict management's operational and financial flexibility. Default can lead to loss of control to creditors.

*   **Dr. Thorne's Concern:** Maintaining control and ensuring rewards for early team is a central theme for founders.

### Linking to Capital Structure Theories

*   **How do these financing choices relate to what we know about capital structure?**

*   **Trade-Off Theory:**
    *   Firms balance tax benefits of debt against costs of financial distress.
    
    *   Young, high-growth firms like MM typically have:
        *   Higher business risk, more volatile earnings (higher distress costs).
        
        *   Fewer tangible assets (less debt capacity).
        
        *   May not be consistently profitable to fully utilize tax shields.
    
    *   Thus, often lean towards more equity financing in early stages. Venture debt is a cautious step towards leverage.

*   **Pecking Order Theory:**
    *   Firms prefer internal funds, then debt, then equity (due to information asymmetry).
    
    *   If MM management (like Dr. Thorne) believes the company is undervalued by outsiders, they would prefer:
        1.  Internal funds (likely insufficient for $50M).
    
        2.  Debt (if accessible and terms are reasonable, e.g., venture debt).
    
        3.  Equity (as a less preferred option, as it might signal the market that current shares are not undervalued).

*   **Signaling:** Choice of financing sends signals to markets.
