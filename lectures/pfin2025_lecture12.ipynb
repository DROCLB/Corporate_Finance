### Introduction to Risk Management

*   **What is Risk?** 
    * Uncertainty about future outcomes that can affect a firm's value or objectives.

*   **Why Manage Risk?**
    *   In perfect markets, risk management by the firm might not add value (investors can diversify on their own - MM Proposition).
    
    *   Value Creation through Market Imperfections:
        *   Reducing costs of financial distress (bankruptcy costs).
        
        *   Minimizing taxes (e.g., by stabilizing earnings).
        
        *   Reducing agency costs (e.g., aligning manager/shareholder interests).
        
        *   Improving investment decisions by ensuring funds are available for positive NPV projects.
        
        *   Reducing information asymmetry for investors.

*   **Risk Management Process:** Identify risks -> Assess risks -> Treat risks (avoid, reduce, transfer, retain) -> Monitor & Review.

*   **Focus today:** Transferring risk (insurance, hedging).

---

### Types of Risk Relevant to Firms

*   Firms face a variety of risks that can impact their operations and financial performance.

*   **Operational Risks:**
    *   Arise from the firm's day-to-day business activities.

    *   Examples:
        *   Supply chain disruptions (e.g., MM's single-source optical sensor).
        *   Product liability (if a robotic arm malfunctions).
        *   Key personnel risk (loss of critical employees like Dr. Thorne).
        *   Cybersecurity breaches.
        *   Natural disasters.

*   **Financial Risks:**
    *   Arise from financial markets and transactions.

    *   Examples:
        *   Exchange rate risk (e.g., MM's CHF payments for sensors).
        *   Interest rate risk (impact on borrowing costs or value of fixed-income assets).
        *   Commodity price risk (for firms reliant on raw materials).
        *   Credit risk (counterparty default).

*   Understanding specific exposures is the first step in management.

### Insurance Fundamentals - Role and Terminology

*   **Insurance:** A contract (policy) where one party (insurer) agrees to indemnify another (insured) against specified losses in return for a payment (premium). It's a risk transfer mechanism.

*   **Key Terminology:**
    *   **Premium:** The fee paid by the insured to the insurer for coverage.
    
    *   **Policy:** The insurance contract detailing terms, coverage, exclusions.
    
    *   **Coverage:** The amount and type of loss protected by the policy.
    
    *   **Deductible:** The amount of loss the insured must pay out-of-pocket before the insurer pays. (Reduces moral hazard, lowers premium).
    
    *   **Policy Limits:** The maximum amount the insurer will pay for a covered loss.

*   **Common Types of Business Insurance:**
    *   Property Insurance (fire, theft, natural disasters).
    
    *   Liability Insurance (general, product, professional).
    
    *   Business Interruption Insurance.
    
    *   Key Personnel Insurance.
    
    *   Specialized policies (e.g., supply chain disruption, cyber risk).

### Insurance Pricing - Actuarially Fair Premium & Cost of Capital

*   **Actuarially Fair Premium (AFP):** In a perfect market, the premium equals the present value of the expected loss.
    *   $AFP = E[Loss] / (1 + r_L)$

    *   $E[Loss] = \text{Probability of Loss} \times \text{Magnitude of Loss}$

*   **Cost of Capital for Insured Loss ($r_L$):**
    *   The appropriate discount rate for the expected loss, reflecting its systematic risk.

    *   Can be estimated using CAPM: $r_L = r_f + \beta_L \times MRP$

    *   $\beta_L$ is the beta of the insured loss (can be zero if diversifiable, or negative if loss is counter-cyclical/pays off in bad times).

    *   Rememeber the beta of financial distress cost?

*   **Example (MM's Sensor Disruption Risk - from case Q7a):**
    *   Prob(Loss) = 5%, Magnitude = $20M, r_f = 3\%, MRP = 7\%, \beta_L = -0.1$.
    
    *   $r_L = 3\% + (-0.1 \times 7\%) = 2.3\%$.
    
    *   $E[Loss] = 0.05 \times \$20M = \$1M$.
    
    *   AFP = $\$1M / (1.023) \approx \$977,517$.

*   Actual premiums include insurer's overhead and profit (loading).

### Value of Insurance - Why Firms Buy It

*   **Beyond AFP:** If insurance is priced fairly (or even with some loading), why do firms buy it? Benefits arise from reducing the impact of market imperfections:
    
    1.  **Reducing Costs of Financial Distress:**
        *   Large uninsured losses can push a firm towards bankruptcy, incurring significant direct (legal) and indirect (lost sales, damaged reputation) costs. Insurance payouts can prevent this.
    
    2.  **Maintaining Optimal Investment:**
        *   Uninsured losses can deplete cash needed for positive NPV projects. Insurance ensures funds are available.
    
    3.  **Tax Benefits:**
        *   Premiums are usually tax-deductible. Tax savings can occur if premiums are paid in high-tax periods and payouts received in low-tax periods (or when firm is loss-making).
    
    4.  **Improving Debt Capacity & Terms:**
        *   Reduces cash flow volatility, making the firm less risky to lenders. Can lead to higher debt capacity or better loan terms (lower interest rates, fewer covenants).
        *   *Relevant for MM considering debt financing.*
    
    5.  **Risk Assessment & Monitoring:** Insurers often have expertise in risk assessment and may provide recommendations for loss prevention.

---

### Costs of Insurance & Market Imperfections

*   While beneficial, insurance isn't free and faces its own market imperfections:
    
    1.  **Administrative Costs (Loading):**
        *   Insurers have operating costs (underwriting, claims processing, sales) and require a profit margin.
        
        *   This means actual premiums > Actuarially Fair Premium.
    
    2.  **Adverse Selection:**
        *   Occurs when the insured party has more information about their risk than the insurer.
        
        *   Higher-risk firms are more likely to buy insurance, potentially driving up average premiums for everyone if insurers cannot perfectly differentiate.
        
        *   Insurers try to mitigate with screening, risk classification.
    
    3.  **Moral Hazard:**
        *   Occurs when having insurance changes the insured's behavior, making them less careful about preventing losses (since the insurer bears the cost).
        
        *   Example: A firm might reduce safety spending after buying fire insurance.
        
        *   Insurers use deductibles, co-payments, policy exclusions, and monitoring to mitigate.

### Exchange Rate Risk - Sources & Exposure

*   What is it?
    * Risk that changes in currency exchange rates will adversely affect a firm's cash flows, profitability, and market value.

*   **Sources of Exposure for Multinational Firms or those with international dealings:**
    
    *   **Transaction Exposure:** Risk on existing contractual cash flows denominated in foreign currency.
    
        *   *MM's payments for optical sensors in CHF are a direct transaction exposure.* If CHF appreciates, MM's USD cost increases.
    
    *   **Translation Exposure (Accounting Exposure):** Risk from converting foreign subsidiaries' financial statements into the parent company's reporting currency. Affects reported earnings, not necessarily cash flows directly.
    
    *   **Economic Exposure (Operating Exposure):** Risk that exchange rate changes will affect a firm's future cash flows and competitiveness, even if it has no direct foreign currency transactions (e.g., due to impact on competitors, input costs, demand).

*   Fluctuations can be significant and unpredictable.

---

### Hedging Exchange Rate Risk - Forward Contracts

*   **Hedging:** Taking an offsetting position in a financial instrument to reduce or eliminate risk.

*   **Currency Forward Contract:**
    *   An agreement to exchange a specific amount of one currency for another at a predetermined future date and a predetermined exchange rate (the forward rate).
    
    *   Customizable contract with a bank.
    
    *   **Mechanism:** Locks in the future exchange rate, eliminating uncertainty.
        *   *If MM needs to pay CHF 5M in one year, it can enter a forward contract today to buy CHF 5M at, say, 0.91 USD/CHF, ensuring its USD cost is fixed at $4.55M regardless of spot rate changes.*

*   **Pricing Intuition (Covered Interest Parity - simplified):**
    *   The forward rate is determined by the spot rate and the interest rate differential between the two currencies. It's set to prevent arbitrage.
    
    *   If $r_{USD} > r_{CHF}$, then USD is expected to depreciate relative to CHF in the forward market (F > S in USD/CHF terms, meaning more USD per CHF).

---

### Hedging Exchange Rate Risk - Currency Options

*   **Currency Option:** Gives the holder the right, but not the obligation, to buy (call option) or sell (put option) a specific amount of currency at a predetermined exchange rate (strike price) on or before a specific date.
    *   Requires paying an upfront premium.

*   **Why use Options instead of Forwards?**
    1.  **Flexibility / Participation in Favorable Moves:**
        *   A forward locks in the rate. An option provides a "worst-case" rate (the strike price + premium) but allows the firm to benefit if the spot rate moves favorably.
        
        *   *If MM buys CHF call options at strike 0.90 USD/CHF and spot rate falls to 0.85, MM can let option expire and buy CHF cheaper at spot. With a forward at 0.91, they'd still pay 0.91.*
    
    2.  **Hedging Contingent Cash Flows:**
        *   If the underlying transaction is uncertain (e.g., a bid for a foreign project), an option is better. If the transaction doesn't happen, the firm only loses the option premium. A forward would still obligate exchange.

*   Options are generally more expensive than forwards due to this flexibility.

---

### Covered Interest Parity (CIP) & Risk Management Strategy

*   **Covered Interest Parity (CIP):** A no-arbitrage condition stating that the forward exchange rate should be such that an investor is indifferent between:
    -  Investing domestically.
    
    -  Investing in a foreign currency and hedging the exchange rate risk with a forward contract.

*   $F = S \times \frac{(1+r_{domestic})}{(1+r_{foreign})}$ (for a given tenor, e.g., 1-year).
    
    *   $S$ = Spot rate, $F$ = Forward rate, $r$ = interest rates.

*   **Arbitrage:** If CIP doesn't hold, risk-free profits are possible.
    
    *   *MM case explores this: S=0.90, F=0.91, $r_{USD}$=3.5%, $r_{CHF}$=1.0%. Theoretical F $\approx$ 0.9223. Market F (0.91) is "too low" for CHF.*

*   **Overall Risk Management Strategy:**
    *   Identify significant risks (like MM's sensor supply & CHF exposure).
    *   Assess their potential impact.
    *   Evaluate mitigation tools (insurance, hedging, operational changes) based on cost vs. benefit.
    *   Consider how risk management impacts financing decisions, investor confidence, and firm value.
    *   It's not about eliminating all risk, but managing it to an acceptable level aligned with strategic objectives.
