### Working Capital Management: The Cash Cycle
* The length of time between when the firm pays cash to purchase inventory and when it receives cash from the sale of the output produced from that inventory.

$$ \text{Cash Conversion Cycle (CCC)} = \text{Inventory Days} + \text{Accounts Receivable Days} - \text{Accounts Payable Days} $$

$$ \text{Inventory Days} = \frac{\text{Inventory}}{\text{Average Daily Cost of Goods Sold}} $$

$$ \text{Accounts Receivable Days} = \frac{\text{Accounts Receivable}}{\text{Average Daily Sales}} $$

$$ \text{Accounts Payable Days} = \frac{\text{Accounts Payable}}{\text{Average Daily Cost of Goods Sold}} $$

### Firm Value and Working Capital
* Any reduction in working capital requirements generates a positive free cash flow that the firm can distribute immediately to shareholders.

* Managing working capital efficiently will maximize firm value.

* Emerald City Paints would like to construct a new facility that will manufacture paint. In addition to the capital expenditure on the plant, management estimates that the project will require an investment today of $450,000 for net working capital. The firm will recover the investment in net working capital eight years from today, when management anticipates closing the plant. The discount rate for this type of cash flow is 6% per year. What is the present value of the cost of working capital for the paint facility? What is the value of an inventory policy that would halve the plant's net working capital requirements?

### Working Capital Management: Trade Credit

* Accounts receivable represent the credit sales for which a firm has yet to receive payment.

* The credit that the firm is extending to its customer is known as trade credit.

* Trade Credit and Market Frictions
    
    * Cost of Trade Credit
        
        * A loan from the selling firm to its customer.
        
        * The price discount represents an interest rate.
        
        * Suppose a firm sells a product for $100 but offers its customer terms of 2/10, Net 30. The customer doesn't have to pay anything for the first 10 days.

If paid in 10 days: $\$100 * (1 – 0.02) = \$98$

If paid in 30 days: $ \frac{100}{98} = (1 + r) \implies r = 2.04\% $

$ EAR = (1 + 0.020408)^{365/20} - 1 = 44.59\% $

### Working Capital Management: Trade Credit
* Can be an attractive source of funds.
    
    * Simple and convenient to use, and therefore has lower transaction costs than alternative sources of funds.
    
    * It is a flexible source of funds, and can be used as needed.
    
    * Sometimes, it is the only source of funding available to a firm.

* Why companies offer trade credit?
    
    * Providing financing at below-market rates is an indirect way to lower prices for only certain customers.
    
    * A supplier may have an ongoing business relationship with its customer, it may have more information about the credit quality of the customer than a traditional outside lender.
    
    * The supplier may also be able to increase the likelihood of payment by threatening to cut off future supplies if payment is not made.
    
    * If the buyer defaults, the supplier may be able to seize the inventory as collateral. This inventory is likely to be more valuable to a company within the industry such as the supplier than to an outsider.

### Working Capital Management: Receivables Management

* Establishing a credit policy involves three steps:
    
    * Establishing credit standards
        
        * The decision of how much credit risk to assume plays a large role in determining how much money a firm ties up in its receivables.
        
        * A restrictive policy can result in a lower sales volume, but the firm will have a smaller investment in receivables.
        
        * A less selective policy will produce higher sales, but the level of receivables will also rise.
    
    * Establishing credit terms
        
        * The length of the period before payment must be made.
        
        * Offer a discount to encourage early payments?
            
            * The discount percentage and the discount period.
    
    * Establishing a collection policy
        
        * What to do when the creditee is late on payments?
            
            * Doing nothing?
            
            * Send a polite letter of inquiry?
            
            * Charge interest on payments extending beyond a specified period?
            
            * Threaten to take legal action at the first late payment?

### Working Capital Management: Payables Management

        * The higher the discount percentage offered, the greater the cost of forgoing the discount.
    
        * The cost of forgoing the discount is also higher with a shorter loan period.
    
    * Monitor the accounts payable, ensure that the payments are made at an optimal time.

    * Stretching the accounts payable.
    
        * Reduces the effective annual rate associated with the trade credit.
    
        * Risks?
            * Cash on delivery (COD) or cash before delivery (CBD).
            * Delinquency

            * Poor credit rating.

### Working Capital Management: Cash Management

* In the Modigliani-Miller setting, the level of cash is irrelevant.

    * With perfect capital markets, a firm is able to raise new money instantly at a fair rate.

    * Also, the firm can invest excess cash at a fair rate to earn an NPV of zero.

* In the real world, markets are not perfect.

    * Liquidity has a cost;
        * Holding liquid assets may earn a below-market return,
        * A firm that needs to raise cash quickly may face transaction costs.

    * Double taxation of corporate interest income.

* The optimal strategy,

    * To hold cash in anticipation of seasonality in operating or investment cash flows,

    * To have buffer for random shocks that affect the business.

### Cash Flow Scenarios: Key Takeaways

This series of scenarios illustrates critical aspects of financial forecasting and cash management:

**Impact of Seasonality:**
*   Even if annual sales and profits are the same as a level-sales scenario, seasonality dramatically impacts intra-year cash flows.

**Sensitivity to Shocks:**
*   **Negative Shock:** Unexpected capital expenditures (equipment failure) directly deplete cash and can turn positive quarterly cash flows negative, highlighting the need for contingency planning or access to credit lines.
*   **Positive Shock:** While increased sales are good, they also require upfront investment in CAPEX and working capital (AR, potentially inventory). The initial cash impact can be negative before the benefits of higher sales are realized.

**Forecasting Horizon & Detail:**
*   Quarterly (or even monthly) forecasting is vital for businesses with seasonality or those undergoing significant changes, as annual averages can mask critical short-term liquidity needs.

**Conclusion for Financial Planning:**

Effective financial planning requires more than just projecting level sales and income. Businesses must:
*   Understand their sales cycle and potential seasonality.
*   Anticipate working capital needs tied to sales fluctuations.
*   Plan for capital expenditures, both routine and unexpected.
*   Stress-test their cash flow forecasts for various shocks.
*   Ensure adequate liquidity (cash reserves, credit lines) to manage troughs in their cash cycle.

### Short-Term Financial Planning: The Matching Principle
* Short-term needs should be financed with short-term debt and long-term needs should be financed with long-term sources of funds.
* Permanent Working Capital
    * The amount that a firm must keep invested in its short-term assets to support its continuing operations.
* Temporary Working Capital
    * The actual level of investment in short-term assets and the permanent working capital investment.
        * Investment in accounts receivable and inventory as a result of seasonal fluctuations in the firm's business or unanticipated shocks.

### Short-Term Financial Planning: Financing Policy Choices

* Aggressive Financing Policy
 
    * Financing part or all of the permanent working capital with short-term debt.
        * Additional transaction costs
        * Interest rate risk
        * Funding risk

* Conservative Financing Policy
    
    * Financing short-term needs with long-term debt.
    
        * Using long-term sources of funds to finance its fixed assets, permanent working capital, and some of its seasonal needs.
    
        * Short-term debt used very sparingly.
    
        * Periods when excess cash is available
    
        * Reduces the funding risk
    
        * Comes with costs
            * Low investment returns on the excess cash
            * Double taxation on the interest income of the excess cash
            * Increased possibility of managers of the firm using the cash nonproductively

### Short-Term Financing with Bank Loans
* Single, End-of-Period Payment Loan
    * Requires that the firm pay interest on the loan and pay back the principal in one lump sum at the end of the loan.
    * The interest rate may be fixed or variable.
* Line of Credit
    * A bank agrees to lend a firm any amount up to a stated maximum.
    * Allows the firm to draw upon the line of credit whenever it chooses.
    * Frequently used to finance seasonal needs.
    * The line of credit may be,
        * An uncommitted line of credit: An informal agreement that does not legally bind the bank to provide the funds.
        * A committed line of credit consists: A legally binding agreement that obligates the bank to provide the funds regardless of the financial condition of the firm, as long as the firm satisfies any restrictions in the agreement.
            * A compensating balance requirement.
            * Restrictions regarding the level of the firm's working capital.
            * A commitment fee based on a percentage of the unused portion of the line of credit.
            * At some point in time the outstanding balance must be zero, which ensures that the firm does not use the short-term financing to finance its long-term obligations.

* Bridge Loans
    * A short-term bank loan that is often used to "bridge the gap” until a firm can arrange for long-term financing.
    * Often quoted as discount loans with fixed interest rates.

### Short-Term Financing with Bank Loans
* Common Loan Stipulations and Fees
    * Commitment Fees
        * A firm negotiates a committed line of credit with a stated maximum of $1 million and an interest rate of 10% (EAR) with a bank. The commitment fee is 0.5% (EAR). At the beginning of the year, the firm borrows $800,000. It then repays this loan at the end of the year, leaving $200,000 unused for the rest of the year.

$ \text{Interest on borrowed funds} = 0.10 * 800,000 = \$80,000 $

$ \text{Commitment fee paid on unused portion} = 0.005 * 200,000 = \$1,000 $

$ \text{Total cost} = \$81,000 $

* Loan Origination Fee.
    * A fee charged by a bank to cover credit checks and legal fees.
    * Paid when the loan is initiated.
    * Reduces the amount of usable proceeds that the firm receives.
    * Timmons Towel and Diaper Service is offered a $500,000 loan for three months at an APR of 12%. This loan has a loan origination fee of 1%, which is charged on the principal of the loan. What is effective annual rate on the loan?

$$ \text{Usable funds} = 500,000 * (1 – 0.01) = \$495,000 $$

$$ \text{Interest on borrowed funds} = 500,000 * \left( \frac{0.12}{4} \right) = \$15,000 $$

$$ EAR = \left( \frac{500,000 + 15,000}{495,000} \right)^4 - 1 = 17.17\% $$

### Short-Term Financing with Bank Loans
* Compensating Balance Requirements.
    * The bank may include a compensating balance requirement in the loan agreement.
    * This reduces the usable loan proceeds.
    * Timmons Towel and Diaper Service is offered a $500,000 loan for three months at an APR of 12%. There is no loan origination fee, but the bank requires the firm to keep an amount equal to 10% of the loan principal in a non-interest-bearing account with the bank as long as the loan remains outstanding. What is effective annual rate on the loan?

$$ \text{Usable funds} = 500,000 * (1 – 0.10) = \$450,000 $$

$$ \text{Interest on borrowed funds} = 500,000 * \left( \frac{0.12}{4} \right) = \$15,000 $$

$$ EAR = \left( \frac{450,000 + 15,000}{450,000} \right)^4 - 1 = 14.01\% $$

### Short-Term Financing with Secured Financing
* Loans collateralized with short-term assets like accounts receivables or inventory.
* Provided by commercial banks, finance companies, and factors.
* Accounts Receivable as Collateral
    * Pledging of Accounts Receivable
        * The lender reviews the invoices that represent the credit sales of the borrowing firm.
        * Notifies which credit accounts it will accept as collateral for the loan.
        * Lends the borrower some percentage of the value of the accepted invoices.
        * The firm is still responsible to the lender for the money.
    * Factoring of Accounts Receivable
        * Firm sells receivables to the lender (i.e., the factor),
        * Lender agrees to pay the firm the amount due from its customers at the end of the firm's payment period.
        * Firm can borrow a percentage of the face value of its receivables from the factor and receive the funds in advance.
            * With recourse: The lender can seek payment from the borrower should the borrower's customers default on their bills.
            * Without recourse: The lender bears the risk of bad-debt losses.
        
---

### Short-Term Financing with Secured Financing
* Inventory as Collateral
    * Floating Lien
        * All of the inventory is used to secure the loan.
        * A risky setup from the standpoint of the lender since the value of the collateral used to secure the loan dwindles as inventory is sold.
        * The loan may become under-collateralized.
    * Trust Receipt
        * Distinguishable inventory items are held in a trust as security for the loan.
        * As these items are sold, the firm remits the proceeds from their sale to the lender in repayment of the loan.
        * The lender checks periodically the inventory items specified in the loan agreement.
    * Warehouse Arrangement
        * The inventory that serves as collateral for the loan is stored in a warehouse.
            * Public warehouse
                * A business that exists for the sole purpose of storing and tracking the inflow and outflow of the inventory.
                * The lender extends a loan to the borrowing firm, based on the value of the inventory stored.
                * Borrowing firm needs get permission from the lender to sell the inventory stored in a warehouse.
            * Field warehouse
                * Operated by a third party but is set up on the borrower's premises in a separate area.
                * Convenient for the borrower.
                * Gives the lender the added security of having the inventory that serves as collateral controlled by a third party.
