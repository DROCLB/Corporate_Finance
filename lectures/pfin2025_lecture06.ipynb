### The Dividend-Discount Model

*   Suppose an investor buys a stock at today's price, $P_0$, to hold for one year, with the expectation that there will be a dividend payment of $D_1$ at t=1. Investor requires an r % return for this stock. In this case, the investor will make the following calculation:
    $$ P_0 = \frac{D_1}{(1+r)} + \frac{P_1}{(1+r)} $$
*   At the end of year 1;
    $$ P_1 = \frac{D_2}{(1+r)} + \frac{P_2}{(1+r)} $$
    $$ P_0 = \frac{D_1}{(1+r)} + \frac{\frac{D_2}{(1+r)} + \frac{P_2}{(1+r)}}{(1+r)} = \frac{D_1}{(1+r)} + \frac{D_2+P_2}{(1+r)^2} = \frac{D_1}{(1+r)} + \frac{D_2}{(1+r)^2} + \frac{P_2}{(1+r)^2} $$
*   If we carry this logic to infinity, then we have the following equation;
    $$ P_0 = \frac{D_1}{(1+r)} + \frac{D_2}{(1+r)^2} + \frac{D_3}{(1+r)^3} + ... = \sum_{t=1}^{\infty} \frac{D_t}{(1+r)^t} $$
*   This is, in fact, a perpetuity. So;
    $$ P_0 = \sum_{t=1}^{\infty} \frac{D_t}{(1+r)^t} = \frac{D}{r} $$

    $$ P_0 = \frac{D}{r} $$

### The Dividend-Discount Model

*   An investor is considering purchasing the shares of AGE Corp. The expected return of the investor is 12%. The company will pay a cash dividend of €3 per share after one year. The dividends paid by the company are expected to grow by 7% for the next 2 years, and then by 5%. What is the maximum price the investor would be willing to pay for this stock?
    $$ P_0 = \underbrace{\frac{3}{(1+0.12)}}_{t_1} + \underbrace{\frac{3(1 + 0.07)}{(1 + 0.12)^2}}_{t_2} + \underbrace{\frac{3(1 + 0.07)^2}{(1 + 0.12)^3}}_{t_3} + \underbrace{\frac{\frac{3(1 + 0.07)^2 * (1 + 0.05)}{0.12 - 0.05}}{(1 + 0.12)^3}}_{t_4} $$
    $$ P_0 = 2.6786 + 2.5590 + 2.4448 + 36.6713 $$
    $$ P_0 = €44.35 $$
*   It is important to predict the dividend growth rate as accurately as possible in the dividend discount model.
*   Firms can be in different growth stages:
    *   **Growth Stage:** Sales are increasing rapidly, high profit margins, high growth in earnings per share, the company has many new investment opportunities, and the dividend payout ratio is low.
    *   **Transition Stage:** Decreased growth rate and profit margin due to competition, fewer new investment opportunities, higher dividend payout rate.
    *   **Maturity Stage:** Stage in which profit growth, dividend payout rate, and average return on investments stabilize over the remaining life of the firm.

### The Dividend-Discount Model

$$ \text{Earnings}_{(t=1)} = \text{Earnings}_{(t=0)} + (\text{Retained Earnings}_0 \times \text{ROE}) $$
Let $NI$ be Net Income, $RE$ be Retained Earnings, and $ROE$ be Return on Equity (on new investments).
$$ NI_1 = NI_0 + (RE_0 \times ROE) $$
Divide by $NI_0$:
$$ \frac{NI_1}{NI_0} = \frac{NI_0}{NI_0} + \frac{RE_0}{NI_0} \times ROE $$
Let $g$ be the growth rate of earnings, so $\frac{NI_1}{NI_0} = (1+g)$.

Let $p$ be the dividend payout ratio, so $\frac{RE_0}{NI_0}$ is the retention ratio $b = (1-p)$.
    
$$ (1 + g) = 1 + (1-p) \times ROE $$
$$ g = (1 - p) \times ROE $$
Also, $p_t = \frac{D_t}{NI_t}$.
$$ \text{If the payout ratio } p \text{ is constant (i.e., } p_0 = p_1 = p \text{), then } \frac{D_1}{D_0} = \frac{p \cdot NI_1}{p \cdot NI_0} = \frac{NI_1}{NI_0} = (1+g) $$

### Growth, Reinvestment, and Return on Invested Capital (ROIC)

*   Consider a business whose operating earnings (NOPAT) for the next year are expected to be the same as this year unless new net investment is made into its core operations.
    *   Net Investment = Capital Expenditures - Depreciation + Change in Non-Cash Working Capital.
    *   This net investment is expected to generate a return, measured by ROIC or RONIC.

    Let NOPAT be Net Operating Profit After Tax and Let $RR$ be the Reinvestment Rate, which is the fraction of NOPAT that is put into Net Investment
    $$ \text{NOPAT}_{(t=1)} = \text{NOPAT}_{(t=0)} + (\text{Net Investment}_0 \times \text{ROIC}) $$
    $$ \text{Net Investment}_0 = RR \times \text{NOPAT}_{(t=0)} \quad \Rightarrow \quad \text{NOPAT}_{(t=1)} = \text{NOPAT}_{(t=0)} + (RR \times \text{NOPAT}_{(t=0)} \times \text{ROIC}) $$
    $$ \frac{\text{NOPAT}_{(t=1)}}{\text{NOPAT}_{(t=0)}} = \frac{\text{NOPAT}_{(t=0)}}{\text{NOPAT}_{(t=0)}} + \frac{RR \times \text{NOPAT}_{(t=0)} \times \text{ROIC}}{\text{NOPAT}_{(t=0)}} $$
    $$ (1 + g) = 1 + (RR \times \text{ROIC}) $$
    $$ g = RR \times \text{ROIC} $$
    Where:
    *   $g$ = Sustainable growth rate of NOPAT (and often FCFF, if payout policies related to NOPAT are stable).
    *   $RR = \frac{\text{Net Investment}}{\text{NOPAT}}$ (Reinvestment Rate from NOPAT).
    *   $ROIC = \frac{\text{NOPAT}}{\text{Invested Capital}}$ (Return on Invested Capital, or RONIC if new investments earn differently.

* A firm's operating growth is driven by how much it reinvests from its operating profits and the return it earns on those new investments.
* If a firm maintains a constant Reinvestment Rate and a constant ROIC, its NOPAT and FCFF will grow at this rate $g$.

*   Alpha Corp. (Α) reported a net profit of \$40 million for the year 2023. The company's board has decided to distribute 45\% of this profit as dividends. Alpha Corp. consistently achieves a ROE of 18\% on its new investments. What is the expected dividend growth rate ($g$) if Alpha Corp. maintains a constant dividend payout ratio?

    $$ g = (\text{Retention Ratio}) \times ROE \quad \Rightarrow \quad g = (1 - \text{Payout Ratio}) \times ROE \quad \Rightarrow \quad g = (1 - 0.45) \times 0.18 \quad \Rightarrow \quad g = 9.9\% $$

*   Alpha Corp. (Α) had 10 million shares outstanding at the end of 2023. Investors require a rate of return of 15% on Alpha Corp.'s shares ($r_E$). Calculate the estimated stock price of the company today ($P_0$) using the constant growth Dividend Discount Model (Gordon Growth Model).

    *   Total Dividends Paid in 2023 = Net Profit $\times$ Payout Ratio $\quad \Rightarrow \quad \$40,000,000 \times 0.45 = \$18,000,000$
    *   Shares Outstanding = $10,000,000$
    *   $D_0 = \frac{\text{Total Dividends}}{\text{Shares Outstanding}} = \frac{\$18,000,000}{10,000,000} = \$1.80 \text{ per share}$

    *   $D_1 = D_0 \times (1+g) \quad \Rightarrow \quad D_1 = \$1.80 \times (1 + 0.099) = \$1.9782 \text{ per share}$

    *   $ P_0 = \frac{D_1}{r_E - g} \quad \Rightarrow \quad P_0 = \frac{\$1.9782}{0.15 - 0.099} \approx \$38.788 $

*   The dividend discount model requires the dividend expected *next period* ($D_1$). Always ensure you are using the forward-looking dividend when $g > 0$.

### The Dividend Discount Model: Components of Return

*   Using the **Constant Growth Dividend Discount Model**, the price of a stock ($P_0$) is:
    $$ P_0 = \frac{D_1}{r-g} \quad \Rightarrow \quad r = \frac{D_1}{P_0} + g $$
    Where $D_1$ is the expected dividend next year, $r$ is the required rate of return, and $g$ is the constant dividend growth rate.

*   **Components of the Expected Total Return ($r$):**

    1.  **Dividend Yield:** $\frac{D_1}{P_0}$
    2.  **Capital Gains Yield (or Price Return):** $g$
        *   In the constant growth model, the stock price is expected to grow at the same rate as the dividends ($g$).
        
        $$P_1 = \frac{D_2}{r-g} = \frac{D_1(1+g)}{r-g} \quad \text{and} \quad \frac{P_1 - P_0}{P_0} = \frac{P_1}{P_0} - 1 \quad \Rightarrow \quad \frac{P_1}{P_0} = \frac{D_1(1+g)/(r-g)}{D_1/(r-g)} = 1+g$$

*   According to the DDM, the expected total return on a stock is the sum of its expected dividend yield and its expected capital gains yield (price appreciation rate):
    $$ \text{Expected Total Return } (r) = \text{Dividend Yield } \left(\frac{D_1}{P_0}\right) + \text{Capital Gains Yield } (g) $$

### The Dividend-Discount Model

Epsilon Dynamics (Ε) reported a net profit of $8,000,000 for the year 2023. The company follows a policy of distributing 40% of its net profit as dividends. Epsilon Dynamics has 2,000,000 shares outstanding. The company's Return on Equity (ROE) is consistently 15%, and its shares are currently trading at $30 in the market.

1.  What is the expected rate of return (required rate of return, $r$) that investors expect when purchasing Epsilon Dynamics stocks?

2.  What is the expected stock price of Epsilon Dynamics after 1 year ($P_1$)?

The sustainable growth rate is the rate at which the company can grow its earnings and dividends without changing its financing policy.
$$ g = \text{Retention Rate} \times ROE = (1 - p) \times ROE \quad \Rightarrow \quad  g = (1 - 0.40) \times 0.15 = 0.60 \times 0.15 = 0.09 = 9\% $$

$$ EPS_0 = \frac{\text{Net Profit}}{\text{Shares Outstanding}} = \frac{\$8,000,000}{2,000,000} = \$4.00  \quad \Rightarrow \quad D_0 = EPS_0 \times p = \$4.00 \times 0.40 = \$1.60 $$

$$ D_1 = D_0 \times (1 + g) = \$1.60 \times (1 + 0.09) = \$1.60 \times 1.09 = \$1.744$$  
$$\quad \text{or} \quad EPS_1 = EPS_0 \times (1 + g) = \$4.00 \times 1.09 = \$4.36  \quad \Rightarrow \quad D_1 = EPS_1 \times p = \$4.36 \times 0.40 = \$1.744$$

$$ r = \frac{D_1}{P_0} + g \quad \Rightarrow \quad r = \frac{\$1.744}{\$30} + 0.09 \quad \Rightarrow \quad r \approx 0.058133 + 0.09 \approx 14.81\% $$

Assuming the growth rate and required return remain constant, the stock price is expected to grow at the same rate as dividends.
$$ P_1 = P_0 \times (1 + g) \quad \Rightarrow \quad P_1 = \$30 \times (1 + 0.09) = \$32.70 $$

**Alternatively, using the DDM for $P_1$:**

$$D_2 = D_1 \times (1 + g) = \$1.744 \times 1.09 = \$1.90096 \quad \Rightarrow \quad P_1 = \frac{D_2}{r - g} = \frac{\$1.90096}{0.148133 - 0.09} \approx \$32.70 $$

### The Dividend Discount Model: Present Value of Growth Opportunities (PVGO)
* Beta Enterprises (Β) has an expected Earnings Per Share ($EPS_1$) of €5.00 for the coming year. The company maintains a dividend payout ratio ($p$) of 40% and earns a Return on Equity (ROE) of 20% on new investments. The required rate of return ($r$) on Beta Enterprises' stock is 15%.

1.  Calculate the expected dividend per share next year ($D_1$).
2.  Calculate the expected growth rate of dividends ($g$).
3.  Calculate the current stock price assuming zero growth ($P_{ng}$), where all earnings are paid out as dividends.
4.  Calculate the current stock price with the given payout ratio and growth ($P_g$).
5.  Determine the Present Value of Growth Opportunities (PVGO).

Given: $EPS_1 = €5.00$, $p = 0.40$ (payout ratio), $ROE = 0.20$, $r = 0.15$, $\; \text{and}$ Retention rate ($b$) = $1 - p = 1 - 0.40 = 0.60$.

**1. Expected Dividend per Share ($D_1$):**
   $D_1 = p \times EPS_1 = 0.40 \times €5.00 = €2.00$

**2. Expected Growth Rate of Dividends ($g$):**
   $g = b \times ROE = (1 - p) \times ROE = 0.60 \times 0.20 = 0.12 \quad (12.0\%)$

**3. Stock Price with Zero Growth ($P_{ng}$):**
   (This implies $p=1$, so $b=0$, thus $g=0$, and $D_1 = EPS_1$)
   $P_{ng} = \frac{EPS_1}{r} = \frac{€5.00}{0.15} = €33.33$

**4. Current Stock Price with Growth ($P_g$):**
   (Using $D_1 = €2.00$ and $g = 0.12$)
   $P_g = \frac{D_1}{r - g} = \frac{€2.00}{0.15 - 0.12} = \frac{€2.00}{0.03} = €66.67$

**5. Present Value of Growth Opportunities (PVGO):**
   $PVGO = P_g - P_{ng}$
   $PVGO = €66.67 - €33.33 = €33.34$

   Alternatively:
   $PVGO = P_g - \frac{EPS_1}{r} = €66.67 - \frac{€5.00}{0.15} = €66.67 - €33.33 = €33.34$

Beta Dynamics (Β)'s earnings and dividends per share are expected to grow indefinitely by 6% a year. Beta Dynamics earns 15% on its new invested capital (ROE = 15%). Next year's dividend ($D_1$) is projected to be $2.00, and the required rate of return on its stock ($r$) is 10%.


$g = 0.06$, $ROE = 0.15$, $D_1 = \$2.00$, $r = 0.10$

$ P_0 = \frac{D_1}{r-g} = \frac{\$2.00}{0.10 - 0.06} = \$50.00 $

$g = (1 - p) \times ROE \quad \Rightarrow \quad 0.06 = (1 - p) \times 0.15 \quad \Rightarrow \quad = 60\%) $

$ D_1 = p \times EPS_1 \implies \$2.00 = 0.60 \times EPS_1 = \$3.3333 $

Price if no growth ($P_{ng}$), where all earnings are paid out $EPS_1$ becomes $D_1$ for a no-growth firm:

$ P_{ng} = \frac{EPS_1}{r} = \frac{\$3.3333}{0.10} = \$33.3333 $

$ PVGO = P_0 - P_{ng} = \$50.00 - \$33.3333 = \approx \$16.67 $

### Share Repurchases and the Total Payout Model
*   An alternative method that may be more reliable when a firm repurchases shares.
*   Values all of the firm's equity
    $$ P_0 = \frac{\text{PV (Future Total Dividends and Repurchases)}}{\text{Shares Outstanding } (t = 0)} $$

* Epsilon Dynamics has 150 million shares outstanding and expects earnings of $600 million at the end of this coming year. Epsilon plans to maintain a total payout ratio of 60% of its earnings. Both earnings and payouts are expected to grow at a constant rate of 5% per year indefinitely. If Epsilon's equity cost of capital is 12%, what is its current share price?

$$ Payout_1 = E_1 \times \text{Total Payout Ratio} \quad \Rightarrow \quad Payout_1 = \$600 \text{ million} \times 0.60 = \$360 \text{ million} $$

$$ PV_{\text{Equity}} = \frac{Payout_1}{r_E - g} \quad \Rightarrow \quad PV_{\text{Equity}} = \frac{\$360 \text{ million}}{0.12 - 0.05} = \frac{\$360 \text{ million}}{0.07} \quad \Rightarrow \quad PV_{\text{Equity}} \approx \$5,142.857 \text{ million} $$

$$ P_0 = \frac{PV_{\text{Equity}}}{\text{Shares Outstanding}} \quad \Rightarrow \quad P_0 = \frac{\$5,142.857 \text{ million}}{150 \text{ million shares}} \approx \$34.2857 \text{ per share} $$

### The Discounted Free Cash Flow Model (DCFM)
*   Discounted cash flow valuation
    *   The value of any asset is the expected cash flows from that asset discounted at a rate that reflects their riskiness (Intrinsic value)
    *   The value of any asset is a function of
        *   the cash flows generated by that asset,
        *   the life of the asset,
        *   the expected growth in the cash flows,
        *   the riskiness associated with these cash flows.
            $$ \text{Value of Asset} = \sum_{t=1}^{N} \frac{E(CF_t)}{(1+r)^t} $$
*   A firm is a portfolio of assets.
*   To value a firm, we can use cash flows to the firm over its life and a discount rate that reflects the collective risk of the firm's assets.
    *   Some of the assets of a firm have already been created.
    *   A significant component of firm value reflects expectations about future investments.
    *   We not only need to measure the cash flows from current investments but also must estimate the expected value from future investments.


---

### The Discounted Free Cash Flow Model (DCFM)
*   Equity Valuation vs. Firm Valuation
    *   Value just the equity stake in the business;
    *   Value the entire firm
    *   Both approaches discount expected cash flows, but the relevant cash flows and discount rates are different for each.
*   The value of equity
    *   Obtained by discounting expected cash flows to equity (the residual cash flows after meeting all operating expenses, tax obligations, and interest and principal payments) at the cost of equity (the rate of return required by equity investors in the firm).
        $$ \text{Value of Equity} = \sum_{t=1}^{N} \frac{E(\text{CF to equity}_t)}{(1+r_e)^t} $$
*   The value of the firm
    *   Obtained by discounting expected cash flows to the firm (residual cash flows after meeting all operating expenses, taxes, and reinvestment needs, but prior to debt payments) using WACC (the cost of the different components of financing used by the firm, weighted by their market-value proportions).
        $$ \text{Value of Firm} = \sum_{t=1}^{N} \frac{E(\text{CF to firm}_t)}{(1 + \text{WACC})^t} $$

### The Discounted Free Cash Flow Model (DCFM)

*   The choice between free cash flows to equity (FCFE) and free cash flows to the firm (FCFF) is a choice between equity and firm valuation.
*   Done consistently, both approaches should yield the same values for the equity in a business.
    *   The cash flows to equity are after net debt issues or payments and become much more difficult to estimate when financial leverage is changing over time.
    *   The cash flows to the firm are pre-debt cash flows and are unaffected by changes in financial leverage. Therefore, firm valuation may be more straightforward.

### The Discounted Free Cash Flow Model (DCFM)

*   Stable growth rate
    *   Because the growth rate in the firm's cash flows is expected to last forever, the firm's other measures of performance (including revenues, earnings, and reinvestment) can be expected to grow at the same rate.
    
    *   In the long run, a firm cannot grow at a rate significantly greater than the growth rate in the economy in which it operates.
        *   If the valuation is done in nominal (real) terms, the stable growth rate cannot be larger than the nominal (real) growth rate in the economy in which the firm operates.
    
    *   Can a stable growth rate be much lower than the growth rate in the economy?
    
    *   Rule of thumb: The stable growth rate should not exceed the risk-free rate used in a valuation. The risk-free rate can be decomposed into an expected inflation rate and an expected real interest rate.
    
    *   If we assume that the real growth rate of an economy will be equal to the real interest rate in the long run, the risk-free rate becomes a proxy for the nominal growth rate in the economy.

### The Discounted Free Cash Flow Model (DCFM)

*   ROE vs ROIC
    *   ROE looks at how much profit a company makes per dollar of shareholders' equity. 
    
    *   ROIC is a measure of how much cash a company gets back for each dollar it invests in its business. 

    $$ROE = \frac{\text{Net Income}}{\text{Book Value of Equity}_{(t-1)}} \quad \text{and} \quad ROIC = \frac{\text{Operating Income} * (1 - \text{tax rate})}{\text{Book Value of Invested Capital}_{(t-1)}} $$
    
    *   Companies that have generated ROICs higher than their WACC for many years running usually have a moat.
    
    *   But a positive spread between ROIC and WACC alone doesn't justify an economic moat.
    
    *   Investors also have to think about the qualitative attributes such as high barriers to entry, huge market share, low-cost production, corporate culture, patents, or high customer switching costs that create an economic moat around a company's profits.

*   Invested Capital 

$$ \text{Invested Capital} = \text{Total Assets} - (\text{Non-Interest-Bearing Current Liabilities}) $$

NIBCLs: Items like Accounts Payable, Accrued Expenses, etc. Spontaneous sources of financing that reduce the need for explicit investor-supplied capital for operations.

$$ \text{NIBCLs} \approx \text{Total Current Liabilities} - \text{Short-term Interest-Bearing Debt} \\
\Rightarrow \quad \text{Invested Capital} = \text{Total Assets} - (\text{Total Current Liabilities} - \text{Short-term Debt}) $$

$$ \text{Invested Capital} = \text{Book Value of Equity} + \text{Book Value of Debt} - \text{Non-Operating Assets (e.g., Excess Cash, Marketable Securities not core to operations)} $$

$$ ROE = ROIC + \frac{\text{Debt}}{\text{Equity}} (\text{ROIC} - \text{After-tax cost of debt}) $$

### The Discounted Free Cash Flow Model - FCFF
*   Allows us to value a firm without explicitly forecasting its dividends, share repurchases, or its use of debt.
*   Enterprise value:
    *   The net cost of acquiring the firm's equity, taking its cash, paying off all debt (owning the unlevered business).
    *   We start with the PV of the free cash flow (FCF) that the firm has available to pay all investors.

    $$ \text{Free Cash Flow} = \underbrace{(\text{EBIT}) \times (1 – \tau_c)}_{\text{Unlevered Net Income}} + \text{Depreciation} – \text{CapEx} – \Delta \text{NWC} $$
    $$ \text{Net Investment} = \text{CapEx} – \text{Depreciation} \quad \Rightarrow \quad \text{Free Cash Flow} = \text{EBIT} \times (1 – \tau_c) – \text{Net Investment} – \Delta \text{NWC} $$
    $$ \text{Enterprise Value} = V_0 = \text{PV (Future Free Cash Flow of Firm)} $$
    $$ \text{Enterprise Value} = \text{Market Value of Equity} + \text{Debt} - \text{Cash} $$
    $$ \text{Market Value of Equity} = \text{Enterprise Value} + \text{Cash} – \text{Debt} $$
    $$ P_0 = \frac{V_0 + \text{Cash}_0 - \text{Debt}_0}{\text{Shares Outstanding}_0} $$

### The Discounted Free Cash Flow Model - FCFF
*   Discount rate
    *   Since we need to discount the free cash flows to all investors of the firm, WACC is the choice.
    
    *   WACC reflects the average risk of all of the firm's investments.
        $$ V_0 = \frac{FCF_1}{(1+r_{WACC})^1} + \frac{FCF_2}{(1+r_{WACC})^2} + ... + \frac{FCF_N + V_N}{(1+r_{WACC})^N} $$
        $$ V_N = \frac{FCF_{N+1}}{r_{WACC} – g_{FCF}} = FCF_N \left( \frac{1+g_{FCF}}{r_{WACC} - g_{FCF}} \right) $$
        Where,
        $V_0$ is the enterprise value
        $V_N$ is the terminal value
        $g_{FCF}$ is the growth rate for FCF beyond the forecast period.
*   Value of company is sum of value from explicit DCF forecast and the discounted continuation (terminal) value.

*   Value of common equity is value of company minus value of claims of investors of higher priority (debt, preferred stock).

*   Capital structure weights
    *   Always use market values if available
    *   Book values of debt may be close if debt is highly rated. Market values often not available.

### The Discounted Free Cash Flow Model - FCFF
*   Valuing a company's equity using enterprise DCF is a four-part process:
    
    1.  Value the company's operations by discounting free cash flow at the weighted average cost of capital.
    
    2.  Identify and value nonoperating assets, such as excess cash and marketable securities, nonconsolidated subsidiaries, and other assets not included in free cash flow. Summing the value of operations and nonoperating assets gives gross enterprise value. 
        
    3.  Identify and value all debt and other nonequity claims against the enterprise value. Debt and other nonequity claims include fixed-rate and floating-rate debt, debt equivalents such as unfunded pension liabilities and restructuring provisions, employee options, preferred stock, among others.

    4.  Subtract the value of debt and other nonequity claims from enterprise value to determine the value of common equity. To estimate value per share, divide equity value by the number of current shares outstanding.
